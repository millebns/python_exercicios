# Faça um Programa que peça o raio de um círculo, calcule e mostre sua área.



## Informações úteis
* Fórmula da área do círculo: pi*r² ou pi*(r*r)
* pi = 3.14

## Algoritmo - Pseudocódigo

Inicio

Variaveis

pi, raio, area

pi <- 3.14

Escreva("Digite o raio do circulo")

Leia raio

area <- pi * (raio * raio)

Escreva("A area do circulo é: ", area)

Fim

## Dicas/Comentários

*   As marcações do Inicio e do Fim são para indicar onde começa e acaba o código, não inserimos estas palavras na programação. 
* O simbolo "<-" indica recebe. Para maiores informações revise a questão 4.
* Na programação utilizamos o ponto para indicar números com vírgulas. Então se você digite 3,14 o programa irá entender como se fosse uma palavra. 
* A ideia é treinar a programação sequencial, por este motivo não inclui nenhum tipo de teste para verificar se o valor digitado é válido. Você poderá criar uma verificação para saber por exemplo se o valor digitado é negativo e solicitar um número novo.

Comandos a serem utilizados:

* input()
* print()

Operadores a serem utilizados:

* float() - utilizamos para converter o valor digitado em um número com vírgula. 
* "*" - operador de multiplicação. 


```python
pi = 3.14
raio = float(input("Digite o raio do circulo "))
area = pi * (raio*raio)
print("A area do circulo é ",area)
```
