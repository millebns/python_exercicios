Comandos a serem utilizados:

* input()
* print()
* format()
** O comando format nos auxilia na formatação do texto a ser exibido na tela. Utilizamos o {} para indicar que iremos substituir os colchetes por uma variável.
** Exemplo: print({} metros é equivalente a {} centimetros".format(m,cm))  Quando for exibir a mensagem na tela, o interpretador encontra os conjuntos de colchetes, {}, solicitando ao método format as variáveis para substituir na frase. Importante: a ordem importa ! Então neste caso o primeiro conjunto de colchetes é substituido pela variável m e o segundo conjunto por cm.


Operadores a serem utilizados:

* float() - utilizamos para converter o valor digitado em um número com vírgula. 
* "+" - operador de soma. 

Observação: 

* explicando o comando float(input("digite o valor ")) 
** o valor digitado pelo usuário será convertido para um número com vírgula. A príncipio, quando você digita um número em Python, o computador entende como se fosse um caractere.

```python
m = float(input("Digite a medida em metros para ser convertido para centimetros "))
cm = m*100
print("{} metros é equivalente a {} centimetros".format(m,cm))
```
