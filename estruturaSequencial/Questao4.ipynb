# Faça um Programa que peça as 4 notas bimestrais e mostre a média.

## Algoritmo - Pseudocódigo

Inicio

Variaveis

primeira_nota, segunda_nota, terceira_nota, quarta_nota, media

Escreva ("Digite a primeira nota ")

Leia primeira_nota

Escreva ("Digite a segunda nota ")

Leia segunda_nota

Escreva ("Digite a terceira nota ")

Leia terceira_nota

Escreva ("Digite a quarta nota ")

Leia quarta_nota

media <- (primeira_nota + segunda_nota + terceira_nota + quarta_nota)/4

Escreva("A média das notas é " media)

Fim

## Dicas/Comentários

*   Neste exercicio, o algoritmo está escrito utilizando o pseudocódigo.
* Não foi incluído nenhum tipo de teste para verificar se as notas são válidas, exemplo se está dentro de 0 a 10. A ideia é treinar as informações sequenciais primeiro, sem nenhum tipo de fluxo. 
*   As marcações do Inicio e do Fim são para indicar onde começa e acaba o código, não inserimos estas palavras na programação. 
* o simbolo "<-" indica recebe, quando escrevemos a <- 2, le-se a variável a recebe o valor 2. 
* o comando Leia, indica que a variável irá receber o valor digitado. 
* Como utilizamos o psudocódigo, declarei as variáveis que vamos utilizar antes.  
* Para deixar o código mais organizado eu coloco um espaço ao final da frase, assim na hora de imprimir e solicitar o valor o número não irá ficar colado na frase. 


Comandos a serem utilizados:

* input()
* print()

Operadores a serem utilizados:

* float() - como o problema fala de notas escolares, e é possível a nota ser um número com vírgula, exemplo: 8,5, vamos utilizar o float. 
* "+" - operador de soma. 

Observação: 

* explicando o comando float(input("digite o valor ")) 
** o valor digitado pelo usuário será convertido para um número com vírgula. A príncipio, quando você digita um número em Python, o computador entende como se fosse um caractere, ou seja, uma palavra/letra.  
** Em outras linguagens de programação, como C e Java você declara a variável explicitando o tipo. 


```python
primeira_nota = 0
segunda_nota = 0
terceira_nota = 0
quarta_nota = 0
media = 0

primeira_nota = float(input("Digite a primeira nota "))
segunda_nota = float(input("Digite a segunda nota "))
terceira_nota = float(input("Digite a terceira nota "))
quarta_nota = float(input("Digite a quarta nota "))
media = (primeira_nota + segunda_nota + terceira_nota + quarta_nota)/4
print("A média das notas é",media)
```
