# Faça um Programa que peça 2 números inteiros e um número real.
# Calcule e mostre:
* ## o produto do dobro do primeiro com metade do segundo .
* ## a soma do triplo do primeiro com o terceiro.
* ## o terceiro elevado ao cubo.


## Algoritmo - Pseudocódigo

Inicio

Variáveis

pri_num: inteiro

seg_num: inteiro

ter_num: real

produto: real

soma: real

cubo: real

Escreva("Digite o primeiro número inteiro:")

Leia pri_num

Escreva("Digite o segundo número inteiro:")

Leia seg_num

Escreva("Digite um número real:")

Leia ter_num

Escreva() // quebra de linha

produto = (pri_num * 2) * (seg_num / 2)

soma = (pri_num * 3) + ter_num

cubo = (ter_num ** 3)

Escreva("Respostas:\n")

Escreva("O produto do dobro do primeiro com metade do segundo é:" produto)

Escreva("A soma do triplo do primeiro com o terceiro é:" soma)

Escreva("O terceiro elevado ao cubo é:" cubo )

Fim


## Dicas/Comentários

* Neste exercício eu mostrarei o como se declara várias variáveis na mesma linha, para isso utilize a vígula como marcador.
*  A ideia é treinar a programação sequencial, por este motivo não inclui nenhum tipo de teste para verificar se o valor digitado é válido. Você poderá criar uma verificação para saber por exemplo se o valor digitado é um número. 
* Iseri a operação dentro de parênteses para deixar o código mais legível. 
* Neste exercício utilizei a notação com o _ para as variáveis
* Aqui eu utilizo dois recursos para deixar o código mais legível, são: 
* print() - para imprimir uma linha em branco, utilizei este recurso para pular uma linha.
* o marcador "\n" - caso você queira pular de linha entre os print, você poderá inserir esse marcador junto ao texto. 

**Atenção:**

* Aqui utilizei o f-string, disponível a partir das versões superiores do Python 3.6, é mais um jeito de formatar as respostas.
* Como utilizar: você deve adicionar a letra "f" antes das aspas e, colocar {variavel} onde deseja inserir o valor.
* Exemplo: print(**f**"Meu nome é **{nome}** e tenho **{idade}** anos")
** Quando quiser formatar as casas decimais: {variavel**:.Xf**} sendo X o número de casas que você deseja depois da vírgula.

Comandos a serem utilizados:

* input()
* print()

Operadores a serem utilizados:

* float() - utilizamos para converter o valor digitado em um número com vírgula. 
* int() - utilizamos para converter o valor digitado em um número inteiro.
* "*" - operador de multiplicação.
* "/" - operador de divisão
* "+" - operador de soma
* "**" - operador para realizar a exponenciação. 



```python
pri_num,seg_num,ter_num = 0,0,0.0
produto,soma, cubo = 0.0,0.0,0.0

pri_num = int(input("Digite o primeiro número inteiro "))
seg_num = int(input("Digite o segundo número inteiro "))
ter_num = float(input("Digite um número real "))
print()
produto = (pri_num * 2) * (seg_num/2)
soma = (pri_num * 3) + ter_num
cubo = (ter_num**3)

print("Respostas: \n")
print(f"O produto do dobro do primeiro com metade do segundo é: {produto:.2f}")
print(f"A soma do triplo do primeiro com o terceiro é: {soma:.2f}")
print(f"O terceiro elevado ao cubo é: {cubo:.2f}")
```
