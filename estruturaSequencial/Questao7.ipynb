# Faça um Programa que calcule a área de um quadrado, em seguida mostre o dobro desta área para o usuário.


## Informações úteis

* A fórmula da área de um quadrado é: lado * lado

## Algoritmo - Pseudocódigo

Inicio

Variaveis
lado, area

Escreva("Digite a medida do lado do quadrado")

Leia lado

area <- lado * lado

Escreva("O valor da area é: ", area, " e o seu dobro é: ", area * 2)

Fim

## Dicas/Comentários

* Aqui quero mostrar que podemos realizar operações dentro da função print, neste caso utilizando o metodo format. 
* A ideia é treinar a programação sequencial, por este motivo não inclui nenhum tipo de teste para verificar se o valor digitado é válido. Você poderá criar uma verificação para saber por exemplo se o valor digitado é negativo e solicitar um número novo.
* Reforço a dica da utilização dos nomes de variáveis claras e descritivas. Para saber sobre boas práticas de programação, revise a questão 2. 

Comandos a serem utilizados:

* input()
* print()
* format()
** O comando format nos auxilia na formatação do texto a ser exibido na tela. Utilizamos o {} para indicar que iremos substituir os colchetes por uma variável.
** Exemplo: print("o primeiro valor é, {} o segundo é {}".format(p,s))  Quando for exibir a mensagem na tela, o interpretador encontra os conjuntos de colchetes, {}, solicitando ao método format as variáveis para substituir na frase. Importante: a ordem importa ! Então neste caso o primeiro conjunto de colchetes é substituido pela variável p e o segundo conjunto por s.


Operadores a serem utilizados:

* float() - utilizamos para converter o valor digitado em um número com vírgula. 
* "*" - operador de multiplicação.

Observação: 

* explicando o comando float(input("digite o valor ")) 
** o valor digitado pelo usuário será convertido para um número com vírgula. A príncipio, quando você digita um número em Python, o computador entende como se fosse um caractere.

```python
lado = float(input("Digite a medida do lado do quadrado "))
area = lado * lado
print("O valor da area é {} e o seu dobro é {}".format(area, area*2))
```
