# Faça um Programa que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês. Calcule e mostre o total do seu salário no referido mês.



## Algoritmo - Pseudocódigo

Inicio

Variaveis

valorHora, horasMes

Escreva("Digite o valor que você ganha por hora")

Leia valorHora

Escreva("Digite o valor da quantidade de horas trabalhadas no mês")

Leia horasMes


Escreva("O valor que irá receber no final do mês é: ",  valorHora * horasMes , "reais")

Fim

## Dicas/Comentários

* Aqui quero mostrar que podemos realizar operações dentro da função print. 
* Iseri a operação dentro de parênteses para deixar o código mais legível. 
* A ideia é treinar a programação sequencial, por este motivo não inclui nenhum tipo de teste para verificar se o valor digitado é válido. Você poderá criar uma verificação para saber por exemplo se o valor digitado é negativo e solicitar um número novo.
* Nesta questão utilizei a notação de camelo para o nomes das variáveis e evite acentos ou caracteres especiais. 

**Comandos a serem utilizados:**

* input()
* print()

**Operadores a serem utilizados:**

* float() - utilizamos para converter o valor digitado em um número com vírgula. 
* "*" - operador de multiplicação.

**Observação:** 

* explicando o comando float(input("digite o valor ")) 
** o valor digitado pelo usuário será convertido para um número com vírgula. A príncipio, quando você digita um número em Python, o computador entende como se fosse um caractere.

```python
valorHora = 0
horasMes = 0

valorHora = float(input("Digite o valor que você ganha por hora "))
horasMes = float(input("Digite o valor da quantidade de horas trabalhadas no mês "))
print("O valor que irá receber no final do mês é ",(valorHora * horasMes), "reais" )
```
