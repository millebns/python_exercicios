# Faça um Programa que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius. 

## Informações úteis

* Fórmula para calcular a temperatura em Celsius com o dado em Fahrenheit: 
* C = 5 * ((F - 32) / 9) 

## Algoritmo - Pseudocódigo

Inicio

Variaveis

tempF, tempC

Escreva("Digite a temperatura em Fahrenheit: ")

Leia tempF

tempC = 5 * ((tempF - 32) / 9)

Escreva("A temperatura em Celsius é: ", tempC, " graus Celsius")

Fim

## Dicas/Comentários

* Neste exercício, foi adicionado comentários ao longo do código. 
* Uma boa prática de programação é a documentação. Tente resumir em uma frase. 
* Como na matemática, podemos utilizar os parenteses para ajudar na priorização das operações. 
* Nesta questão, utilizo a função format para auxiliar na exibição da resposta com apenas 2 casas decimais.
* Como fazer a restrição das casas decimais com o format:
* 1. Dentro das chaves {}, você irá inserir :.Xf - o X representa o número de casas que você deseja. 
* 2. Exemplo: para imprimir uma variável com três casas decimais print({:.3f}.format(variavel))

Comandos a serem utilizados:

* input()
* print()
* format()

Operadores a serem utilizados:

* float() - utilizamos para converter o valor digitado em um número com vírgula. 
* "*" - operador de multiplicação.
* "/" - operador de divisão

```python
# Declaração das variáveis
tempF = 0
tempC = 0

# Solicita ao usuário para inserir a temperatura em Fahrenheit

tempF = float(input("Digite a temperatura em Fahrenheit "))

# Calcula a temperatura em Celsius usando a fórmula (tempF - 32) * 5/9

tempC = 5 * ((tempF -32)/9)

# Imprime o valor da temperatura em Celsius arredondada para duas casas decimais

print("A temperatura em Celsius é: {:.2f}".format(tempC))
```
