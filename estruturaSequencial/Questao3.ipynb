# Faça um Programa que peça dois números e imprima a soma

# Algoritmo:

variavel primeiro_numero recebe (Escreva("Digite o primeiro número "))

variavel segundo_numero recebe (Escreva("Digite o segundo número "))

variavel soma recebe primeiro_numero + segundo_numero 

Escreva("O resultado dos dois números digitados é" soma)

Operadores a serem utilizados:

* float() - aqui nós reforçamos que o número digitado será um número real, caso queira trabalhar somente com números inteiros utilize o int()
* "+" - o operadora realiza a operação de soma caso todas as variaveis envolvidas sejam números (precisa garantir que tudo seja número), caso não seja número o operador irá fazer a concatenação das variaveis. 

Exemplo 1:

Exemplo 2:

```python
primeiro_numero = float(input("Digite o primeiro número "))
segundo_numero = float(input("Digite o segundo número "))
soma = primeiro_numero + segundo_numero
print("O resultado dos dois números digitados é {}".format(soma))
```
