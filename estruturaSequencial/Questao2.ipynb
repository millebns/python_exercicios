## Dicas/Comentários: 

Lembre-se existem algumas boas práticas de programação com os nomes de variáveis, seguem algumas sugestões:

* iniciar com a letra minúscula
* não conter acentos ou qualquer caracter especial, exceto o _ 
* colocar nomes que facilitam o entendimento do que está sendo trabalhado
* aderir um padrão de nomenclatura
** Exemplos: notacaoDeCamelo ou utilizacao_underscore

Comandos a serem utilizados:
* input()
* print()

Para realizar a junção de um texto com uma variavel no comando print existem algumas opções:
* print("texto" + variavel)
* print("texto { }".format(variavel)) 
** nesta opção utilizamos o método format, este método é bastante útil quando temos mais de uma variavel ou queremos formartar o resultado. Exemplo: exibir um número com duas casas decimais.

```python
numero = input("Digite um número ")
print("O número digitado foi:{}".format(numero))
```
