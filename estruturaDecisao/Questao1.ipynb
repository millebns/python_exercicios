# Faça um Programa que peça dois números e imprima o maior deles.


##Algoritmo - Peusocódigo

1. Solicitar ao usuário para digitar o primeiro número.

2. Armazenar o valor digitado em uma variável chamada "primeiro_numero" como um número de ponto flutuante.

3. Solicitar ao usuário para digitar o segundo número.

4. Armazenar o valor digitado em uma variável chamada "segundo_numero" como um número de ponto flutuante.

5. Verificar se o "primeiro_numero" é maior ou igual ao "segundo_numero".

6. Se a condição for verdadeira, imprimir "O maior número é: " seguido pelo valor de "primeiro_numero".

7. Caso contrário, imprimir "O maior número é: " seguido pelo valor de "segundo_numero".

## Dicas/Comentários


*   A ideia é introduzir o conceito de condicional, por este motivo não vou incluir nenhum tipo de teste para verificar se o valor digitado é válido.
* Antes de realizar a comparação dos números, é feita a conversão dos valores inseridos pelo usuário para o tipo float usando a função float(). Isso permite lidar com números decimais e garantir que a comparação seja feita corretamente.
*   Neste exercicio vou apresentar a estrutura do IF ELSE
```
if (condicional):
  faça esta ação
else: 
  faça esta ação
```

* Podemos traduzir o IF e ELSE para SE e SENÃO, ou seja, SE a nossa CONDIÇÂO estiver correta, ele irá fazer a ação abaixo. SENÂO ele pula para a ação abaixo do else.
* Iseri a condicional dentro de parênteses para deixar o código mais legível.

**Atenção:**

* Não se esqueça de acrescentar o **:** no final da condicional e ao lado do else.
* A organização do código é super importante, devemos indentar corretamente.

**Comandos a serem utilizados:**

* input()
* print()
* if(condição):
* else:

**Operadores a serem utilizados:**

* float() - utilizamos para converter o valor digitado em um número com vírgula. 
* ">=" - operador maior e igual.
* Opcional: Se você quisesse identificar apenas o número estritamente maior, seria necessário usar o operador ">" no lugar de ">=".
** ">" - operador maior que 






````python
primeiro_numero = float(input("Digite o primeiro número "))
segundo_numero = float(input("Digite o segundo número "))
if (primeiro_numero >= segundo_numero ):
    print("O maior número é: ",primeiro_numero)
else:
    print("O maior número é: ",segundo_numero)
````
