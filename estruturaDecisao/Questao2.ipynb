#Faça um Programa que peça um valor e mostre na tela se o valor é positivo ou negativo



##Algoritmo - Pseudocódigo
1. Ler numero como um número de ponto flutuante
2. Se numero < 0 então
     3. Imprimir "O valor é negativo"
     4. Senão Imprimir "O valor é positivo"
5. Fim


## Dicas/Comentários

*   A ideia é introduzir o conceito de condicional, por este motivo não vou incluir nenhum tipo de teste para verificar se o valor digitado é válido.
* Antes de realizar a comparação dos números, é feita a conversão dos valores inseridos pelo usuário para o tipo float usando a função float(). Isso permite lidar com números decimais.
* Para realizar este exercicio, utilizei a comparação número digitado < 0, para consultar se o número é negativo.
*   Neste exercicio eu utilizei a estrutura do IF-ELSE
```
if (condicional):
    faça esta ação
else: 
    faça esta ação
```
* Quando utilizamos o if-else, o programa verifica a nossa condição para saber qual comando que ele irá executar.
* SE (IF) a condição for verdadeira, ele executa a próxima linha. 
* SENÂO (ELSE), o programa executa o que está embaixo do else, ignorando tudo que esta escrito entre o if e else. 

**Atenção:**

* Não se esqueça de acrescentar o **:** no final da condicional e ao lado do else.
* A organização do código é super importante, devemos indentar corretamente.

**Comandos a serem utilizados:**

* input()
* print()
* if(condição):
* else:

**Operadores a serem utilizados:**

* float() - utilizamos para converter o valor digitado em um número com vírgula. 
* "<" - operador menor que



````python
numero = float(input("Digite um número "))
if numero < 0: 
  print("O valor é negativo")
else:
  print("O valor é positivo")
````
