#Faça um programa que verifique se uma letra digitada é vogal ou consoante.






##Algoritmo - Pseudocódigo 
1. Criar uma lista chamada "vogais" contendo as vogais ['a','e','i','o','u', 'A','E','I','O','U']
2. Ler a letra do usuário e armazenar em uma variável chamada "letra"
3. Se letra estiver presente na lista vogais então
     4. Imprimir "é vogal"
     5. Senão Imprimir "é consoante"
6. Fim


## Dicas/Comentários

* Neste exercicio crio uma lista com todas as vogais.
* Como no Python, letras maiusculas são diferentes de letras minusculas, devemos acrescentar na lista as vogais em maiusculas e minusculas. 
* A ideia do exercicio é consultar se a letra digitada pertence a lista de vogais.
* Iseri a condicional dentro de parênteses para deixar o código mais legível.

**Atenção:**

* Não se esqueça de acrescentar o **:** no final da condicional e ao lado do else.
* A organização do código é super importante, devemos indentar corretamente.

**Comandos a serem utilizados:**

* input()
* print()
* if(condição):
* else:

**Operadores a serem utilizados:**

* "in" - para consultar se algo pertence, ou seja, está dentro da nossa lista criada. 

**Explicação sobre lista**

* As listas são uma estrutura de dados que armazena uma coleção de elementos.
* A declaraçõ da lista é: nome = [item 1, item 2]
* Caso a sua lista seja um conjunto de letras ou palavras, você deve colocar os itens dentro de aspas simples ou duplas.
* Caso a sua lista seja um conjunto de número, não precisa utilizar as aspas.
* Podemos guardar diferentes tipos de dados dentro de uma lista.
* Utilizamos a vírgula para separar os itens dentro da lista.


Exemplos:

* numeros = [1, 2, 3, 4, 5]
* nomes = ["João", "Maria", "Pedro", "Ana"]
* dados = [10, "abc", True, 3.14]



```python
vogais = ['a','e','i','o','u', 'A','E','I','O','U']

letra = input("Digite uma letra ")
if (letra in vogais):
  print("é vogal")
else:
  print("é consoante")
```
